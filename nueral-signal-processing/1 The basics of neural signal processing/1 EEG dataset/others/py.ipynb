```python
import numpy as np
import scipy.io as sio
from pytopo import topoplotIndie
```

```python
matfile = sio.loadmat("../../../data/sampleEEGdata")
EEG = matfile["EEG"][0, 0]
erp = EEG["data"].mean(2)
```

```python
i = np.argmin(abs(EEG["times"][0] - 300))
EEG["times"][0][i], i
```

```python
def get_timepoint(t):
    i = np.argmin(abs(EEG["times"][0] - t))
    return i


get_timepoint(300)
```

```python
topoplotIndie(erp[:, get_timepoint(100)], EEG["chanlocs"])
topoplotIndie(erp[:, get_timepoint(300)], EEG["chanlocs"])
```
